```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("/content/Electric_Vehicle_Population_Data.csv")
df.head()
```

# Most vehicle registered in which year?

```python
ev_adoption_by_year = df['Model Year'].value_counts().sort_index()
sns.barplot(x=ev_adoption_by_year.index, y=ev_adoption_by_year.values)
plt.title('EV Adoption Over Time')
plt.xlabel('Model Year')
plt.ylabel('Number of Vehicles Registered')
plt.xticks(rotation=60)
plt.show()
```

# Top 7 Counties in which most EV are registered

```python
ev_adoption_by_county=df['County'].value_counts().head(7)
sns.barplot(x=ev_adoption_by_county.values,y=ev_adoption_by_county.index )
plt.title('Top Counties by EV Registrations')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('County')
plt.tight_layout()
plt.show()
```

# Top 10 City where EV was registered the most

```python
ev_adoption_by_city=df['City'].value_counts().head(10)
sns.barplot(x=ev_adoption_by_city.values,y=ev_adoption_by_city.index )
plt.title('Top Cities in Top Counties by EV Registrations')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

# Now lets find out which brand cars were registered most in Seattle.


```python
seattle_df=df[df['City']=='Seattle']
seattle_car_registered=seattle_df['Make'].value_counts().head(10)
sns.barplot(x=seattle_car_registered.values,y=seattle_car_registered.index )
plt.title('Top Brands in Seattle by EV Registrations')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Brand')
```

# Types of EV registered the most

```python
ev_registered_type=df['Electric Vehicle Type'].value_counts()
sns.barplot(y=ev_registered_type.values,x=ev_registered_type.index )
plt.title('Top Electric Vehicle Type by EV Registrations')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Electric Vehicle Type')
```

# Which company's most EV was registered in the United States.

```python
ev_registered_make=df['Make'].value_counts().head(10)
sns.barplot(x=ev_registered_make.values,y=ev_registered_make.index,color='yellow')
plt.title('Top Make by EV Registrations')
plt.xlabel('Number of Vehicles Registered')
```

# Lets Find out in Tesla what model was registered the most

```python
Tesla_registered_model=df[df['Make']=='TESLA']
Tesla_registered_model=Tesla_registered_model['Model'].value_counts()
sns.barplot(y=Tesla_registered_model.values,x=Tesla_registered_model.index ,color='red')
plt.title('Top Tesla Model by EV Registrations')
```

# Distribution of electric ranges

 Identify any notable trends, such as improvements over time or variations between different vehicle types or manufacturers:

```python
sns.histplot(df['Electric Range'], bins=30, kde=True, color='green')
plt.title('Distribution of Electric Vehicle Ranges')
plt.xlabel('Electric Range (miles)')
plt.ylabel('Number of Vehicles')
plt.axvline(df['Electric Range'].mean(), color='red', linestyle=':', label=f'Mean Range: {df["Electric Range"].mean():.2f} miles')
plt.legend()
plt.show()
```

# Average electric Range by model year

```python
average_range_by_year = df.groupby('Model Year')['Electric Range'].mean().reset_index()

sns.lineplot(x='Model Year', y='Electric Range', data=average_range_by_year, marker='o', color='red')
plt.title('Average Electric Range by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Average Electric Range (miles)')
plt.grid(True)
plt.show()
```
